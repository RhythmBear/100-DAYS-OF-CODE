# Data Visualisation

### Notebook Formatting & Style Helpers

```python

# Create locators for ticks on the time axis

# Create Axis and set plot size
plt.figure(figsize=(14,8), dpi=120)
ax1 = plt.gca()
ax2 = ax1.twinx()

# Set and Customize Axes
ax1.set_ylabel('TSLA STOCK PRICE', color="#E6232E", fontsize=12)
ax2.set_ylabel("Search Trend", color="skyblue", fontsize=12)

# Plot graph and add settings
ax1.plot(df_tesla.MONTH, df_tesla.TSLA_USD_CLOSE,  color="#E6232E", linewidth=3)
ax2.plot(df_tesla.MONTH, df_tesla.TSLA_WEB_SEARCH, color="skyblue", linewidth=3)

# Create Title
plt.title('TESLA WEB SEARCH VS PRICE', pad=20)
plt.show()
```

```python
# Register date converters to avoid warning messages
## Creating the matplotlib.dates object for sectioning the month axis
# Create a Year locator

# Create year locator object 
years = mdates.YearLocator()

# Create month locator object
months = mdates.MonthLocator()

# Create years Formatter
years_format = mdates.DateFormatter('%Y')

# Add objects to plotter ax1
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_format)
ax1.xaxis.set_minor_locator(months)

# Plot graph and add settings
ax1.plot(df_tesla.MONTH, df_tesla.TSLA_USD_CLOSE,  color="#E6232E", linewidth=3)
ax2.plot(df_tesla.MONTH, df_tesla.TSLA_WEB_SEARCH, color="skyblue", linewidth=3)
```

### Unemployement Benefits Search vs. Actual Unemployment in the U.S.

**Challenge** Plot the search for "unemployment benefits" against the unemployment rate. 
1. Change the title to: Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate <br>
2. Change the y-axis label to: FRED U/E Rate <br>
3. Change the axis limits <br>
4. Add a grey [grid](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.grid.html) to the chart to better see the years and the U/E rate values. Use dashes for the line style<br> 
5. Can you discern any seasonality in the searches? Is there a pattern? 

```python
# dataframe = df_unemployment
df_unemployment.head()

plt.figure(figsize=(14,7), dpi=120)
plt.title("Monthly Search of 'unemployment benefits' against the Unemloyment Rate", fontsize=12, pad=20)


#CREate date markers
years = mdates.YearLocator()
months = mdates.MonthLocator()
year_fmt = mdates.DateFormatter('%Y')


# Define axes
ax1 = plt.gca()
ax2 = ax1.twinx()
ax1.grid(color='grey', linestyle='--')

# Edit yaxis labels
ax1.set_ylabel("FRED U/E Rate", color='purple', fontsize=12)
ax2.set_ylabel("SEARCH TRENDS", color='skyblue', fontsize=12)

# SET LOCATORS FOR X AXIS
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(year_fmt)
ax1.xaxis.set_minor_locator(months)


#Set axes limits
ax1.set_xlim(left=df_unemployment.MONTH.min(), right=df_unemployment.MONTH.max())
ax1.set_ylim(bottom=2, top=11)
ax2.set_ylim(bottom=0, top=100)

# CALCUlate rolling window
roll_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()

# SET GRID
#plot graph
ax1.plot(df_unemployment.MONTH, roll_df.UNRATE, color='purple', linestyle='--' )
ax2.plot(df_unemployment.MONTH, roll_df.UE_BENEFITS_WEB_SEARCH, color='skyblue')


```
